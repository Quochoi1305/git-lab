### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/u.user).

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/u.user',
                   sep='|')

display(users.head())
```

### Step 4. Discover what is the mean age per occupation

```python
mean_age_per_occupation = users.groupby('occupation')['age'].mean()
display(mean_age_per_occupation)
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
def gender_to_numeric(gender):
    if gender == 'M':
        return 1
    else:
        return 0

users['gender_numeric'] = users['gender'].apply(gender_to_numeric)
male_ratio_per_occupation = users.groupby('occupation')['gender_numeric'].mean().sort_values(ascending=False)
display(male_ratio_per_occupation)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
min_max_age_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])
display(min_max_age_per_occupation)
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
mean_age_per_occupation_gender = users.groupby(['occupation', 'gender'])['age'].mean()
display(mean_age_per_occupation_gender)
```

### Step 8.  For each occupation present the percentage of women and men

```python
gender_counts = users.groupby(['occupation', 'gender']).size()
occupation_gender_percentage = gender_counts.groupby(level=0).apply(lambda x: 100 * x / x.sum())

display(occupation_gender_percentage.unstack(fill_value=0))
```
