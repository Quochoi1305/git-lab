### Step 3. Use the tsv file and assign it to a dataframe called food

```python
import pandas as pd
import kagglehub
import os

# Download latest version
path = kagglehub.dataset_download("openfoodfacts/world-food-facts")
print("Path to dataset files:", path)

# Construct the full path to the tsv file
tsv_file_path = os.path.join(path, "en.openfoodfacts.org.products.tsv")

# Read the tsv file into a dataframe
food = pd.read_csv(tsv_file_path, sep="\t")
```
