### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
url = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv'
chipo = pd.read_csv(url, sep = '\t')
display(chipo.head())
```

### Step 4. How many products cost more than $10.00?

```python
# Clean the item_price column to remove '$' and convert to float (already done in previous run, but keeping for idempotency)
# chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype(float)

# Filter the original DataFrame for items costing more than $10.00
items_over_10_df = chipo[chipo['item_price'] > 10.
unique_items_over_10 = items_over_10_df['item_name'].nunique()

print(f"Number of unique products costing more than $10.00: {unique_items_over_10}")
```

### Step 5. What is the price of each item?
###### print a data frame with only two columns item_name and item_price

```python

item_and_price = chipo[['item_name', 'item_price']]
unique_item_and_price = item_and_price.drop_duplicates()
unique_item_and_price = unique_item_and_price.sort_values(by='item_name')
display(unique_item_and_price)
```

### Step 6. Sort by the name of the item

```python

display(unique_item_and_price)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python

max_price = chipo['item_price'].max()


most_expensive_items = chipo[chipo['item_price'] == max_price]


total_quantity_most_expensive = most_expensive_items['quantity'].sum()

print(f"The quantity of the most expensive item(s) ordered is: {total_quantity_most_expensive}")
```
