### 7. (3đ) Tạo pie chart thể hiện phần trăm của cột race và sex (sẽ có 5*2=10 categories tổng thể trong 1 hình vẽ). Nhận xét xem nhóm nào chiếm tỉ lệ nhiều nhất.

```python
df['race_sex'] = df['race'] + ' - ' + df['sex']

race_sex_counts = df['race_sex'].value_counts()

plt.figure(figsize=(12, 12))
plt.pie(race_sex_counts, labels=race_sex_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Phân phối chủng tộc và giới tính kết hợp (Race and Sex Distribution)')
plt.axis('equal') # Ensures that pie is drawn as a circle.
plt.show()

print('Nhóm chiếm tỷ lệ nhiều nhất:')
print(race_sex_counts.idxmax(), f'({race_sex_counts.max()} người, chiếm {race_sex_counts.max() / len(df) * 100:.1f}%)')
```
